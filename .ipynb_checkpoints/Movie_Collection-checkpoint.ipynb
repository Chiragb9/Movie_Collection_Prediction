```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./Datasets/Movie_regression.csv", header=0)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

# Null Value Imputation

```python
df["Time_taken"].fillna(df["Time_taken"].mean(), inplace=True)
```

```python
df.info()
```

# Dummy Variables

```python
df=pd.get_dummies(df, columns=["3D_available","Genre"], drop_first=True)
```

```python
df.head()
```

```python
X=df.loc[:,df.columns!="Collection"]
X.shape
```

```python
y=df["Collection"]
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_test.shape
```

```python
from sklearn import tree
```

```python
reg_tree = tree.DecisionTreeRegressor(max_depth=3)
```

```python
reg_tree.fit(X_train, y_train)
```

```python
y_test_pred=reg_tree.predict(X_test)
```

```python
y_train_pred=reg_tree.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

# Plotting the Tree

```python
dot_tree = tree.export_graphviz(reg_tree, out_file=None, feature_names=X_train.columns,filled=True)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_tree)
```

```python
Image(graph.create_png())
```

# Tree Pruning

# Maximum Depth

```python
reg_tree_1 = tree.DecisionTreeRegressor(max_depth=3)
reg_tree_1.fit(X_train, y_train)
dot_tree_1 = tree.export_graphviz(reg_tree_1, out_file=None,feature_names=X_train.columns,filled=True)
graph_1 = pydotplus.graph_from_dot_data(dot_tree_1)
Image(graph_1.create_png())
```

# Minimum Observation at internal node

```python
reg_tree_2 = tree.DecisionTreeRegressor(min_samples_split=40)
reg_tree_2.fit(X_train, y_train)
dot_tree_2 = tree.export_graphviz(reg_tree_2, out_file=None,feature_names=X_train.columns,filled=True)
graph_2 = pydotplus.graph_from_dot_data(dot_tree_2)
Image(graph_2.create_png())
```

# Minimum Observation at leaf node

```python
reg_tree_3 = tree.DecisionTreeRegressor(min_samples_leaf=25)
reg_tree_3.fit(X_train, y_train)
dot_tree_3 = tree.export_graphviz(reg_tree_3, out_file=None,feature_names=X_train.columns,filled=True)
graph_3 = pydotplus.graph_from_dot_data(dot_tree_3)
Image(graph_3.create_png())
```
